```python
import os
from collections import defaultdict

def scan_folder_and_count_types(folder_path):
    file_counts = defaultdict(int)  # 默认的文件计数器
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            file_extension = file.split('.')[-1].lower() if '.' in file else 'no_extension'
            file_counts[file_extension] += 1
    return dict(file_counts)

folder_path = 'C:/Users/Dingz/Desktop/273'  # 替换为你的文件夹路径
file_counts = scan_folder_and_count_types(folder_path)

# 打印统计结果
for file_type, count in file_counts.items():
    print(f"{file_type.upper()} 文件: {count} 个")
```

```python
from PyPDF2 import PdfReader

def extract_pdf_text(pdf_path):
    try:
        reader = PdfReader(pdf_path)
        text = ""
        for page in reader.pages:
            text += page.extract_text()
        return text[:1000]  # 提取前1000个字符
    except Exception as e:
        return f"Error extracting text from PDF: {str(e)}"

import nbformat

def extract_ipynb_text(ipynb_path):
    try:
        with open(ipynb_path, 'r', encoding='utf-8') as f:
            notebook = nbformat.read(f, as_version=4)
            text = ""
            for cell in notebook.cells:
                if cell.cell_type == 'code' or cell.cell_type == 'markdown':
                    text += ''.join(cell['source'])
            return text[:1000]  # 提取前1000个字符
    except Exception as e:
        return f"Error extracting text from IPYNB: {str(e)}"
    
def extract_py_text(py_path):
    try:
        with open(py_path, 'r', encoding='utf-8') as f:
            return f.read()[:1000]  # 读取前1000个字符
    except Exception as e:
        return f"Error extracting text from PY: {str(e)}"
```

```python
import os

def scan_folder_and_extract_content(folder_path):
    file_data = {}
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            file_path = os.path.join(root, file)
            file_extension = file.split('.')[-1].lower()

            if file_extension == 'pdf':
                file_data[file_path] = {"type": "pdf", "content": extract_pdf_text(file_path)}
            elif file_extension == 'ipynb':
                file_data[file_path] = {"type": "ipynb", "content": extract_ipynb_text(file_path)}
            # elif file_extension == 'csv':
            #     file_data[file_path] = {"type": "csv", "content": extract_csv_text(file_path)}
            # elif file_extension == 'txt':
            #     file_data[file_path] = {"type": "txt", "content": extract_txt_text(file_path)}
            elif file_extension == 'py':
                file_data[file_path] = {"type": "py", "content": extract_py_text(file_path)}
            else:
                file_data[file_path] = {"type": "other", "content": "No text extracted"}
    
    return file_data

# 扫描文件夹并提取内容
folder_path = 'C:/Users/Dingz/Desktop/273'  # 替换为你的文件夹路径
file_data = scan_folder_and_extract_content(folder_path)
```

```python
# 使用ollama生成目录
import ollama
from tqdm import tqdm
import time

def generate_summary_with_ollama(text):
    try:
        # 使用 Ollama 模型生成内容摘要
        response = ollama.chat(model="qwen2.5:14b", messages=[{"role": "user", "content": text}])
        return response['text']
    except Exception as e:
        return f"Error generating summary: {str(e)}"

# 逐文件处理并保存到txt文件
def process_files_and_save_to_txt(folder_path, output_txt_file):
    # 获取所有文件
    file_paths = []
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            file_paths.append(os.path.join(root, file))

    # 打开文件进行写入
    with open(output_txt_file, 'w', encoding='utf-8') as f:
        # 使用tqdm显示进度条
        for file_path in tqdm(file_paths, desc="Processing Files", unit="file"):
            file_extension = file_path.split('.')[-1].lower()
            content = ""

            if file_extension == 'pdf':
                content = extract_pdf_text(file_path)
            elif file_extension == 'ipynb':
                content = extract_ipynb_text(file_path)
            # elif file_extension == 'csv':
            #     content = extract_csv_text(file_path)
            # elif file_extension == 'txt':
            #     content = extract_txt_text(file_path)
            elif file_extension == 'py':
                content = extract_py_text(file_path)
            else:
                content = "No text extracted"

            # 生成摘要并写入文件
            summary = generate_summary_with_ollama(content)
            f.write(f"File: {file_path}\nSummary: {summary}\n\n")
            f.flush()

# 示例：使用函数处理文件夹
folder_path = 'C:/Users/Dingz/Desktop/273'  # 替换为你的文件夹路径
output_txt_file = os.path.join(folder_path, "file_directory_summary.txt")  # 保存目录摘要的文件路径
process_files_and_save_to_txt(folder_path, output_txt_file)
```
